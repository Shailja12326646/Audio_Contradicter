```python
import os, whisper

audio_dir = "/content/extracted_data/INNOV8 3.0/INNOV8 3.0"
model = whisper.load_model("medium")  # better accuracy with noisy audio

transcripts = {}

for file in os.listdir(audio_dir):
    if file.endswith(".mp3") or file.endswith(".wav"):
        result = model.transcribe(os.path.join(audio_dir, file))
        transcripts[file] = result["text"]
        print(f"✅ Transcribed: {file}")
```

```python
import nltk
nltk.download("punkt")

def to_sentences(text):
    return nltk.sent_tokenize(text)
```

```python
from transformers import pipeline

nli = pipeline("text-classification", model="roberta-large-mnli")

def contradiction_pairs(texts_by_session):
    sentences = {s: to_sentences(text) for s, text in texts_by_session.items()}
    contradictions = []

    sess = list(sentences.keys())
    for i in range(len(sess)):
        for j in range(i+1, len(sess)):
            for s1 in sentences[sess[i]]:
                for s2 in sentences[sess[j]]:
                    result = nli(f"{s1} </s> {s2}")
                    if result[0]["label"] == "CONTRADICTION":
                        contradictions.append((s1, s2))
    return contradictions
```
